#Sequential Search pada Data Angka

```python
def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return f"Data (target) ditemukan di indeks {i}"
  return f"Data (target) tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

#Binary Search pada Data Angka

```python
def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] ==  target:
      return f"Data {target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

#Sequential Search untuk Data Nama

```python
def sequential_search_nama(data, target):
  for i in range(len(data)):
    if data[i].lower() == target.lower():
      return f"Nama {target} ditemukan di indeks {i}"
  return f"Nama {target} tidak ditemukan"

nama = ["Andi", "Budi", "Cici", "Doni"]
print("Data:", nama)
print(sequential_search_nama(nama, "Cici"))
print(sequential_search_nama(nama, "Eka"))
```

#Studi Kasus Nilai MAhasiswa dengan Binary Search

```python
def binary_search_nilai(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"Nilai {target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"Nilai {target} tidak ditemukan"

nilai = [60, 70, 75, 80, 85, 90]
print("Data nilai:", nilai)
print(binary_search_nilai(nilai, 85))
print(binary_search_nilai(nilai, 65))
```

#Perbandingan Waktu Sequential vs Binary Search

```python
import time

def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return i
  return -1

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return mid
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return -1

data = list(range(1, 100000))
target = 99999

start = time.time()
sequential_search(data, target)
end = time.time()
print("Sequential Search waktu:", end - start, "detik")

start = time.time()
binary_search(data, target)
end = time.time()
print("Binary Search waktu:", end - start, "detik")
```

#Kombinasi Sorting dan Searching

```python
import random

def insertion_sort(data):
  for i in range(1, len(data)):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"{target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"{target} tidak ditemukan"

data = [random.randint(1, 100) for _ in range(15)]
print("Data acak sebelum diurutkan:", data)
urut = insertion_sort(data)
print("Data setelah diurutkan:", urut)
target = urut[7] #ambil angka ke-8 sebagai target
print(binary_search(urut, target))
```

#Tugas 3: Variasi Sequential Search

```python
def sequential_search_count(data, target):
  count = 0
  for i in range(len(data)):
    count += 1
    if data[i] == target:
      return f"Data ditemukan di indeks {i}, perbandingan: {count}"
  return f"Data tidak ditemukan, perbandingan: {count}"

data = [10, 20, 30, 40, 50]
print(sequential_search_count(data, 30))
print(sequential_search_count(data, 100))
```

#Tugas 4: Variasi Binary Search

```python
def binary_search_count(data, target):
  left, right = 0, len(data)-1
  count = 0
  while left <= right:
    count += 1
    mid = (left + right) // 2
    if data[mid] == target:
      return f"Ditemukan di indeks {mid}, iterasi: {count}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"Data tidak ditemukan, iterasi: {count}"

data = [10, 20, 30, 40, 50]
print(binary_search_count(data, 40))
print(binary_search_count(data, 25))
```

#Tugas 6: Studi Kasus Barang Toko

```python
def binary_search_barang(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid][1] == target:
      return f"Barang ditemukan: {data[mid]}"
    elif data[mid][1] < target:
      left = mid + 1
    else:
      right = mid - 1
  return "Barang tidak ditemukan"

barang = [("Buku", 5000), ("Tas", 80000), ("Sepatu", 20000)]
barang.sort(key=lambda x: x[1])
print(binary_search_barang(barang, 80000))
print(binary_search_barang(barang, 50000))
```

#Tugas 7: Analisis Kompleksitas

```python
import time

def seq(data, target):
  for x in data:
    if x == target:
      return

def bin_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1

sizes = [100, 1000, 10000]
print("Ukuran | Sequential | Binary")
for size in sizes:
  data = list(range(size))
  target = size - 1

  start = time.time()
  seq(data, target)
  t1 = time.time() - start

  start = time.time()
  bin_search(data, target)
  t2 = time.time() - start

  print(size, "|", t1, "|", t2)
```
