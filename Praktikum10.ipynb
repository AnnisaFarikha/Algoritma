#Bubble Sort

```python
def bubble_sort(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

#Selection Sort

```python
def selection_sort(data):
    n = len(data)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if data[j] < data[min_index]:
                min_index = j
        data[i], data[min_index] = data[min_index], data[i]
    return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

#Insertion Sort

```python
def insertion_sort(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

#Sorting Data Barang (Nama & Harga)

```python
def bubble_sort_key(data, key_index=0, reverse=False):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if reverse:
                if data[j][key_index] < data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
            else:
                if data[j][key_index] > data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
    return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

#Tugas 3: Eksperimen Bubble Sort

```python
def bubble_sort_desc(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] < arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

data = [25,12,90]
print("Sebelum:", data)
print("Descending:", bubble_sort_desc(data.copy()))
```

#Tugas 4: Eksperimen Selection Sort

```python
def selection_sort_place_max_end(arr):
    n = len(arr)
    for i in range(n-1):
        max_idx = 0
        for j in range(1, n - i):
            if arr[j] > arr[max_idx]:
                max_idx = j
        arr[max_idx], arr[n - i - 1] = arr[n - i - 1], arr[max_idx]
    return arr

data = [22,64,11]
print("Before:", data)
print("After:", selection_sort_place_max_end(data.copy()))
```

#Tugas 5: Eksperimen Insertion Sort

```python
def insertion_sort_strings(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j].lower() > key.lower():
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr

names = ["Nisa","Aini","Kyla"]
print("Before:", names)
print("Sorted:", insertion_sort_strings(names.copy()))
```

#Tugas 7: Analisis Kompleksitas

```python
import random
def bubble_count(arr):
    n = len(arr); count = 0
    a = arr.copy()
    for i in range(n):
        for j in range(0, n-i-1):
            count += 1
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
    return count

def selection_count(arr):
    n = len(arr); count = 0
    a = arr.copy()
    for i in range(n-1):
        max_idx = 0
        for j in range(1, n-i):
            count += 1
            if a[j] > a[max_idx]:
                max_idx = j
        a[max_idx], a[n-i-1] = a[n-i-1], a[max_idx]
    return count

def insertion_count(arr):
    n = len(arr); count = 0
    a = arr.copy()
    for i in range(1,n):
        key = a[i]; j = i-1
        while j >= 0:
            count += 1
            if a[j] > key:
                a[j+1] = a[j]
                j -= 1
            else:
                break
        a[j+1] = key
    return count

for n in (10,50,100):
    data = random.sample(range(n*5), n)
    print("n=", n)
    print("bubble:", bubble_count(data))
    print("selection:", selection_count(data))
    print("insertion:", insertion_count(data))
    print("----")
```

#Tugas 8: Proyek Mini - Manajemen Barang

```python
def bubble_sort_key(data, key_index=0, reverse=False):
    n = len(data)
    a = data.copy()
    for i in range(n):
        for j in range(0, n - i - 1):
            left = a[j][key_index]
            right = a[j+1][key_index]
            if reverse:
                if left < right:
                    a[j], a[j+1] = a[j+1], a[j]
            else:
                if left > right:
                    a[j], a[j+1] = a[j+1], a[j]
    return a

barang = [("Buku",15000),("Pensil",5000),("Tas",75000),("Pulpen",10000)]
print("Urut nama asc:", bubble_sort_key(barang, key_index=0, reverse=False))
print("Urut harga desc:", bubble_sort_key(barang, key_index=1, reverse=True))
```
