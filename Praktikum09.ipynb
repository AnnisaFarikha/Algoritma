#Tugas 3: Ekspresi dengan Modul math

```python
import math

def luas_lingkaran(r):
    return math.pi * r ** 2

print("Luas lingkaran jari-jari 7:", luas_lingkaran(7))
print("Luas lingkaran jari-jari 10:", luas_lingkaran(10))
```

#Tugas 6: Modul Buatan Sendiri

```python
import konversi

print("25°C =", konversi.celcius_ke_fahrenheit(25), "°F")
print("77°F =", konversi.fahrenheit_ke_celcius(77), "°C")
```

```python
##KONVERSI.PY
def celcius_ke_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_ke_celcius(f):
    return (f - 32) * 5/9
```

#Tugas 8: Studi Kasus Mini

```python
import random
import laporan

benar = 0
total = 3

for i in range(total):
    a = random.randint(1, 10)
    b = random.randint(1, 10)
    jawaban = int(input(f"{a} + {b} = "))
    if jawaban == a + b:
        benar += 1

laporan.tampilkan_skor(benar, total)
```

```python
##laporan.py
def tampilkan_skor(benar, total):
    print(f"Skor akhir: {benar}/{total}")
    print("Terima kasih sudah bermain!")
```
