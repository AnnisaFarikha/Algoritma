#Fungsi dengan Default Argument

```python
def sapa(nama="Mahasiswa"):
    print(f"Halo, {nama}!")

sapa()        #tanpa argumen
sapa("Andi")  #dengan argumen
```

#Fungsi dengan *args dan **kwargs

```python
def jumlahkan(*args):
    return sum(args)

def biodata(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

#Scope Variabel

```python
x = 10

def ubah():
    x = 5
    print("Nilai lokal:", x)

ubah()
print("Nilai global:", x)
```

#Fungsi Lambda

```python
kali = lambda x, y: x * y
print("Hasil perkalian:", kali(4,5))
```

#Fungsi Decorator

```python
def decorator_sapa(fungsi):
    def wrapper():
        print("Sebelum fungsi dijalankan...")
        fungsi()
        print("Setelah fungsi dijalankan...")
    return wrapper

@decorator_sapa
def sapa():
    print("Halo, Mahasiswa Polines!")

sapa()
```

#Studi Kasus: Menghitung Rata-rata Nilai Mahasiswa

#Tugas 5: Fungsi dengan Default Argument

```python
def salam(waktu="pagi", nama="Mahasiswa"):
    print("Selamat", waktu, ",", nama)

salam()
salam("siang")
salam("malam", "Budi")
```

#Tugas 6: *Fungsi dengan args

```python
def jumlahkan_semua(*angka):
    return sum(angka)

print(jumlahkan_semua(1, 2, 3))
print(jumlahkan_semua(4, 5, 6, 7, 8))
print(jumlahkan_semua(10, 20, 30, 40, 50, 60, 70))
```

#Tugas 7: **Fungsi dengan kwargs

```python
def info_barang(**data):
    for key, value in data.items():
        print(key, ":", value)

info_barang(Nama="Buku", Harga=15000, Stok=30)
```

#Tugas 8: Scope Variabel

```python
x = 100

def ubah():
    x = 50
    print("Nilai lokal:", x)

ubah()
print("Nilai global:", x)
```

#Tugas 9: Fungsi Lambda

```python
kuadrat = lambda x: x ** 2
keliling = lambda r: 2 * 3.14 * r

print("Kuadrat:", kuadrat(5))
print("Keliling lingkaran:", keliling(7))
```

#Tugas 10: Fungsi Decorator

```python
def logger(func):
    def wrapper():
        print("Memulai fungsi...")
        func()
        print("Fungsi selesai dijalankan.")
    return wrapper

@logger
def halo():
    print("Halo, Mahasiswa Polines!")

halo()
```

#Tugas 11: Studi Kasus Nyata - Program Nilai

```python
def input_nilai():
    return [80, 75, 90, 60, 85]

def hitung_rata(data):
    return sum(data) / len(data)

def status_kelulusan(rata):
    if rata >= 60:
        return "Lulus"
    else:
        return "Tidak Lulus"

nilai_akhir = lambda tugas, ujian: (0.4 * tugas) + (0.6 * ujian)

data = input_nilai()
rata = hitung_rata(data)
status = status_kelulusan(rata)
akhir = nilai_akhir(80, 90)

print("Rata-rata:", rata)
print("Status:", status)
print("Nilai akhir (berbobot):", akhir)
```
